In statistics, exploratory data analysis is an approach of analyzing data sets to summarize their main characteristics, often using statistical graphics and other data visualization. 
EDA helps to spot any unusual data or outliers and is usually done before starting more detailed statistical analysis or building models. 

```python
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
data = pd.read_csv('dataset.csv')  # Replace with your dataset file path

# Preview the dataset
print(data.head())
```

Overview of the Dataset

```python
# Dataset overview
print("Shape of the dataset:", data.shape)
print("\nColumn Information:\n")
print(data.info())

# Check for missing values
print("\nMissing Values:\n", data.isnull().sum())
```

```python
# Example: Countplot for 'LUNG_CANCER'
sns.countplot(x='LUNG_CANCER', data=data, palette='viridis')
plt.title('Lung Cancer Distribution')
plt.show()
```

Univariate Analysis : analyzing one variable at a time.

```python
# Plot distributions for numerical features
sns.histplot(data['AGE'], bins=15, kde=True, color='blue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()

# Bar charts for categorical variables
for col in categorical_columns[:-1]:  # Exclude target variable
    sns.countplot(x=col, data=data, palette='viridis')
    plt.title(f'{col} Distribution')
    plt.show()
```

 Bivariate Analysis : It involves the analysis of two variables (often denoted as X, Y), for the purpose of determining the empirical relationship between them.

```python
# Boxplot: Age vs Lung Cancer
sns.boxplot(x='LUNG_CANCER', y='AGE', data=data, palette='coolwarm')
plt.title('Age vs Lung Cancer')
plt.show()

# Categorical variables vs Lung Cancer
for col in categorical_columns[:-1]:  # Exclude target variable
    sns.countplot(x=col, hue='LUNG_CANCER', data=data, palette='viridis')
    plt.title(f'{col} vs Lung Cancer')
    plt.show()
```

Correlation Analysis is statistical method that is used to discover if there is a relationship between two variables/datasets, and how strong that relationship may be.

```python
# Compute correlation matrix
correlation_matrix = data.corr(numeric_only=True)

# Visualize correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(data, hue='LUNG_CANCER', palette='coolwarm')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Example: Enlarging a scatter plot
plt.figure(figsize=(12, 8))  # Width = 12, Height = 8 (in inches)
plt.scatter(x=data['AGE'], y=data['ALCOHOL CONSUMING'], color='blue')
plt.title("Age vs Alcohol Consuming", fontsize=16)  # Increase font size
plt.xlabel("Age", fontsize=14)
plt.ylabel("Alcohol Consuming", fontsize=14)
plt.grid(True)
plt.show()
```
